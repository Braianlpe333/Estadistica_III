```python
# 2.1.2 Para las variables numericas crear histrogramas. Listar los modelos de carros que estan mas lejos de 4 estandares
#       de desviacion, y serian considerados outliers. Hacer test de si es una distribucion normal o no.
```

```python
col = [                 
    "Invoice",        
    "EngineSize",     
    "Cylinders",      
    "Horsepower",      
    "MPG_City",         
    "MPG_Highway",      
    "Weight",           
    "Wheelbase",        
    "Length"           
]
```

```python
df_stand = (df[col]-df[col].mean())/df[col].std()
df_stand.query("(Invoice > 4.0 or Invoice < -4.0)"
               "or (EngineSize > 4.0 or EngineSize < -4.0)"
               "or (Cylinders > 4.0 or Cylinders < -4.0)"
               "or (Horsepower > 4.0 or Horsepower < -4.0)"
               "or (MPG_City > 4.0 or MPG_City < -4.0)"
               "or (MPG_Highway > 4.0 or MPG_Highway < -4.0)"
               "or (Weight > 4.0 or Weight < -4.0)"
               "or (Wheelbase > 4.0 or Wheelbase < -4.0)"
               "or (Length > 4.0 or Length < -4.0)")
```

```python
df_exclude = df_stand.query("(Invoice < 4.0 or Invoice > -4.0)"
               "or (EngineSize < 4.0 or EngineSize > -4.0)"
               "or (Cylinders < 4.0 or Cylinders > -4.0)"
               "or (Horsepower < 4.0 or Horsepower > -4.0)"
               "or (MPG_City < 4.0 or MPG_City > -4.0)"
               "or (MPG_Highway < 4.0 or MPG_Highway > -4.0)"
               "or (Weight < 4.0 or Weight > -4.0)"
               "or (Wheelbase < 4.0 or Wheelbase > -4.0)"
               "or (Length < 4.0 or Length > -4.0)")
df_exclude
```

```python
shapiro_test = stats.shapiro(df_exclude)
shapiro_test 
```

```python
shapiro_test = stats.shapiro(df[col])
shapiro_test 
```

```python
#Segun el test de shapiro-Wilk la distribucion tanto con los datos excluidos como no excluidos nos muestra una distribucion normal
#Esto debido a que la poca cantidad de outliers encontrados no nos dan un cambio sustancial en los estadisticos
```
