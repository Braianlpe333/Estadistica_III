```python
import pandas as pd
import math
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
#Copiamos el codigo dado en la tarea y se calculan las derivadas parciales para el gradiente descendiente
```

```python
def y_predict(a,b,c,d,x):
    return a * np.sin(b*x + c) + d

def gradient_descent(x, y, lr, batch, epochs):
	n = len(x)
	rsl = []
	a = random.random()
	b = random.random()
	c = random.random()
	d = 0
	for i in range(epochs):
		a_gradiente = 0
		b_gradiente = 0
		c_gradiente = 0
		d_gradiente = 0
		error = 0
		e=0
		for m in range(batch):
			ix = int(random.uniform(0,n))
			y_predic = y_predict(a,b,c,d,x[ix])
			error = y[ix] - y_predic
			e += error**2
			a_gradiente += 2 * error * -1.0*(np.sin(b * x[ix] + c))
			b_gradiente += 2 * error * -1.0*(np.cos(b * x[ix] + c) * x[ix])
			c_gradiente += 2 * error * -1.0*(np.cos(b * x[ix] + c))
			d_gradiente += 2 * error
            
		a -= lr * a_gradiente/batch
		b -= lr * b_gradiente/batch
		c -= lr * c_gradiente/batch
        
		error = error/batch
		rsl.append([a,b,c,error])
		print(f"error:{e} period:{b} amplitude:{a} constant:{c}")
	return rsl, a, b, c
```

```python
#Importamos los datos de la se√±al dada por la tarea
```

```python
data = pd.read_csv("senal.csv")
```

```python
#y los separamos tanto en x como en y para poder hacer la aproximacion
```

```python
x = data["x"].values
y = data["y"].values
```

```python
#Empezamos entonces a evaluar con cada uno de los Epochs Solicitados:
```

```python
#Con 10 Epochs
```

```python
lr = 0.01
batch = 500
epochs = 10

resultados, a, b, c = gradient_descent(x,y,lr, batch,epochs)
```

```python
plt.ylabel("y")
plt.plot(y)
ix=-1
plt.plot([y_predict(a,b,c,0,xi) for xi in x[:ix]])
```
