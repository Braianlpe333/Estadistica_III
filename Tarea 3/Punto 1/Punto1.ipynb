```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from matplotlib import pyplot as plt
from sklearn import tree
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
%matplotlib inline
```

```python
#Primero importamos un data frame creado a mano para los datos ofrecidos por la tarea 3
```

```python
df = pd.read_excel("Punto1JN.xlsx")
```

```python
df.info()
```

```python
#Creamos las variables a usar donde la variable objetivo es el credito
```

```python
x = df.drop(['Credito'], axis='columns')
y = df['Credito']
```

```python
#Dividimos en test y train
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.8, random_state = 0)
```

```python
#Creamos el random forest con los datos de train
```

```python
arbol = tree.DecisionTreeClassifier(random_state=0,criterion='entropy')

arbol_credito = arbol.fit(x_train, y_train)
```

```python
#Graficamos el arbol
```

```python
fig = plt.figure(figsize=(40,40))

tree.plot_tree(arbol_credito, feature_names=list(x.columns.values),
              class_names=['No Credito','Credito'], filled=True)
plt.show()
```
